```python
import csv
import yfinance as yf
```

# CHANGE LABEL TO BE +1 date to account for next day effect!!

```python
dates = []
y = []

with open('Combined_News_DJIA.csv') as csv_file:
    with open('processed_data.csv', 'w') as csv_file1:
        csv_writer = csv.writer(csv_file1, delimiter=',')
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                csv_writer.writerow(row)
                line_count += 1
            else:
                try:
                    data = yf.download('GBPUSD=X',row[0],row[0])
                    open_price = data.iloc[0]['Open']
                    close_price = data.iloc[0]['Close']
                    result = close_price - open_price
                    if result < 0:
                        y.append('1')
                    else:
                        y.append('0')
                    row[1] = y[line_count - 1]
                    csv_writer.writerow(row)
                except:
                    y.append('0')
                    row[1] = y[line_count - 1]
                    csv_writer.writerow(row)
                line_count += 1
    # TOTAL OF 1990 days of data times 25 news stories each day
```
